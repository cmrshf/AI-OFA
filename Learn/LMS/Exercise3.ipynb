```python
# konversi bilangan desimal ke biner
# konversi bilangan desimal ke oktal
# konversi bilangan desimal ke heksadesimal
```

```python
# program konversi bilangan sederhana
# untuk membantu memahami terkait bilangan biner, oktal, & heksadesimal
# program konversi bilangan desimal berbasis 10

number = int(input("Masukan bilangan berbasis 10\n"))

# pilih konversi
print("a. Desimal menjadi Heksadesimal")
print("b. Desimal menjadi Oktal")
print("c. Desimal menjadi Biner")

# memasukkan input
print("Pilih Konversi : - ")
choise = input()

# variable pemilihan
# kondisi
# heksadesimal dipilih jika "a"

if choise is 'a':
  # lstrip helps remove "0x" from the left
  # rstrip helps remove "L" from the right,
  # L represents a long number
  print("Heksadesimal dari bilangan " + str(number) + " adalah " + hex(number).lstrip("0x").rstrip("L"))

if choise is 'b':

  # Octal representation is done
  # by adding a prefix "0o"
  print("Oktal dari bilangan " + str(number) + " adalah c"+bin(number).lstrip("0o").rstrip("L"))

if choise is 'c':

  # Binary representation is done by
  # the addition of prefix "0b"
  print("Biner dari bilangan " + str(number) + " adalah "+bin(number).lstrip("0b").rstrip("L"))
```
